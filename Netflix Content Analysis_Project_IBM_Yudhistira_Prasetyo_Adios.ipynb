```python
# Import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import os
from google.colab import userdata

# Setup API
api_token = userdata.get("api_token")
os.environ["REPLICATE_API_TOKEN"] = api_token

# Import LLM
from langchain_community.llms import Replicate

# Setup parameter LLM
parameters = {
  "top_k": 5,
  "top_p": 1.0,
  "max_tokens": 4096,
  "min_tokens": 0,
  "random_seed": None,
  "repetition_penalty": 1.0,
  "temperature": 0.7,
  "stopping_criteria": "length (256 tokens)",
  "stopping_sequence": None
}

llm = Replicate(
    model="ibm-granite/granite-3.3-8b-instruct",
    input=parameters,
)
# Load Dataset

df = pd.read_csv('netflix_titles.csv')
df.head()
```

## Movie vs TV Show

```python
# Distribusi Movie vs TV Show
plt.figure(figsize=(6,6))
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#d81f25','#7b7b7b'])
plt.title('Distribusi Tipe Konten')
plt.ylabel('')
plt.show()
```

## Number of releases per year

```python
plt.figure(figsize=(12, 6))
df['release_year'].value_counts().sort_index().plot(
    kind='bar',
    color='#d81f25'
)
plt.title("Jumlah Konten Dirilis per Tahun")
plt.xlabel("Tahun")
plt.ylabel("Jumlah")
plt.grid(True)
plt.show()
```

## Most countries

```python
top_countries = df['country'].value_counts().head(10)

top_countries.plot(
    kind='barh',
    color='#4a4a4a'
)

plt.title("Top 10 Countries of Origin for Most Content")
plt.xlabel("Jumlah")
plt.ylabel("Negara")
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## Highest ranking

```python
df['rating'].value_counts().plot(
    kind='bar',
    color='#d81f25'
)

plt.title("Highest ranking")
plt.xlabel("Rating")
plt.ylabel("Amount")
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## Genre WordCloud

```python
from wordcloud import WordCloud

all_genres = ','.join(df['listed_in'])
custom_colors = [ '#d81f25','#4a4a4a', '#f07191', '#ffffff']

import random
def custom_color_func(*args, **kwargs):
    return random.choice(custom_colors)

# Generate WordCloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    color_func=custom_color_func,
    collocations=False
).generate(all_genres)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Wordcloud Genre', fontsize=16)
plt.tight_layout()
plt.show()
```
